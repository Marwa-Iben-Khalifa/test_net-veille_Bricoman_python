```python
# Importing BeautifulSoup class from the bs4 module
from bs4 import BeautifulSoup
# Importing requets module
import requests
# Importing csv module
import csv
  
# Opening the html file
HTMLFile = open("data.html", "r")
  
# Reading the file
index = HTMLFile.read()
  
# Creating a BeautifulSoup object and specifying the parser
S = BeautifulSoup(index, 'lxml')
  
#print(s)
urls=S.find_all("a", class_="product-card-container")
print(urls)
```

```python
def product_obj(obj):
    dataObj = [];
    #soup = BeautifulSoup(obj, 'lxml')
    dataObj.append("https://www.bricoman.fr"+obj["href"])
    dataObj.append(obj.find("p", class_="product-name").get_text())
    dataObj.append(obj.find("p", class_="col-11").get_text().split("\n")[2])
    dataObj.append(''.join(obj.find("div", class_="ttcommons-bold").get_text().split(' ')))
    dataObj.append(obj.find("img", class_="q-mb-xs")["src"])
    dataObj.append(obj.find("span", class_="product-ref").get_text())
    return dataObj

productList=[]
for url in urls:
    productList.append(product_obj(url))
print(productList)
```

```python
#définir une lite de headers pour le fichier csv
headers=["product Url","product Title", "product Brand","product Price", "product Image Url","product Ref"]
#exporter les résultats dans un fichier CSV
with open('scrapedData.csv', 'w') as f:
    # using csv.writer method from CSV package
    write = csv.writer(f)
      
    write.writerow(headers)
    write.writerows(productList)
```
